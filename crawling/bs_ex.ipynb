```python
from bs4 import BeautifulSoup as bs
#HTML 분석해서 원하는 데이터만 뽑기
```

```python
#응답

#url = "https://expample.com/news"
#res = requests.get(url)
#html = res.text

html = '''<h1 id="title">한빛출판네트워크</h1><div class="top"><ul
class="menu"><li><a href="http://www.hanbit.co.kr/member/login.html"
class="login">로그인 </a></li></ul><ul class="brand"><li><a href="http://www.hanbit.co.kr/media/">한빛미디어<li><a href="http://www.hanbit.co.kr/academy/">한빛아카데미</a></li></ul></div>'''
```

```python
# text ->html DOM : 태그단위로 접근 가능
soup = bs(html, 'html.parser')
print(soup.prettify()) # 저장된 내용 확인
```

```python
#태그명으로 태그 찾기
soup.h1
```

```python
#여러개 있으면 맨 첫번째 것 1개만 찾는다
tag_a = soup.a
print(tag_a)
```

```python
#없는 태그를 찾으면 None
soup.h2
```

```python
#find() 하나 찾기
tag_a = soup.find('a')
print(tag_a)
```

```python
#find_all() 여러개 찾기
tag_a = soup.find_all('a')
for a in tag_a : print(a)
```

```python
#attrs : {태그의 속성 이름 : 속성값} 형태의 dict로 출력
soup.h1.attrs
```

```python
soup.find('a').attrs
```

```python
#특정 속성 값 추출
for a in tag_a:
    print(a['href'])
```

```python
#ul 태그의 class 이름이 brand를 찾는다
brand = soup.find('ul', attrs={'class':'brand'})
print(brand)
```

```python
title = soup.find(id='title')
print(title)
```

```python
#select() CSS 선택자로 찾기
li_list = soup.select('div > ul.brand > li')
print (li_list)
```

```python
#태그의 text 가져오기.
for li in li_list:
    print(li.text)
```
