```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
#크롤링 차단을 피하기 위한 기본 세팅
#기본 requests 요청 ->봇으로 인식됨 ->차단
#User_Agent : 나는 브라우저입니다!

header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Trident/7.0; rv:11.0) like Gecko'}
req = requests.get('https://www.melon.com/chart/index.htm', headers = header) ## 주간 차트를 크롤링 할 것임

#response 받기
html = req.text #응답데이터, 단순 str 상태

#browser처럼 document로 변환
parse = bs(html, 'html.parser')

#노래 제목, 가수, 앨범이름 저장 공간 선언
title = []
singer = []
album = []
t_titles = parse.find_all('div', {'class':'ellipsis rank01'})

for t in t_titles:
    title.append(t.find('a').text)
print(title)

t_singers = parse.find_all('div', {'class' : 'ellipsis rank02'})
t_albums = parse.find_all('div', {'class' : 'ellipsis rank03'})
 
for t  in t_titles:
    title.append(t.find('a').text)
 
for s in t_singers:
    singer.append(s.find('span', {'class':'checkEllipsis'}).text)
 
for a  in t_albums:
    album.append(a.find('a').text)

RANK = 100
 
for i in range(RANK):
    print('%3d위: %s [ %s ] - %s'%(i+1, title[i], album[i], singer[i]))
```
