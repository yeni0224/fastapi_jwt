```python
#hollys에서 1~n페이지에 있는 매장 정보 크롤링 후 csv로 저장
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
def hollys_store(result):
    for page in range(1,46):
        h_url = 'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo=%d&sido=&gugun=&store=' %page
        req = requests.get(h_url)
        html = req.text
        soup_hollys = bs(html, 'html.parser')

        #tbody 태그 찾기
        tag_body = soup_hollys.find('tbody')
        #tr 태그 여러개
        for store in tag_body.find_all('tr'):
            #td[0] td[1] td[2] td[3] td[5]
            store_td= store.find_all('td')
            s_name = store_td[0].text
            s_sido = store_td[1].text
            s_address = store_td[3].text
            s_phone = store_td[5].text

            result.append([s_name]+[s_sido]+[s_address]+[s_phone])
    return
        

def main():
    result=[]
    print('start hollys crawlling>>>>>>>>>>>>>>>>>>>>>>>>>>>>>')
    hollys_store(result)
    print ('end hollys crawling>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>')
    hollys_tbl = pd.DataFrame(result, columns=('store','si-do-gu','address','phone'))
    hollys_tbl.to_csv('hollys.csv', mode='w', index=True)#w:기존파일내용덮어쓰기

if __name__ =='__main__':
    main()
```
